```python
from getpass import getpass
import os

REPLICATE_API_TOKEN = getpass()
os.environ["REPLICATE_API_TOKEN"] = REPLICATE_API_TOKEN
```

```python
from llama_index.core import Settings, VectorStoreIndex
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.llms.replicate import Replicate

Settings.llm = Replicate(model="meta/meta-llama-3-8b-instruct",temperature=0.01, additional_kwargs={"top_p": 1, "max_new_tokens": 500},)

Settings.embed_model = HuggingFaceEmbedding( model_name="BAAI/bge-small-en-v1.5")
```

<font color=Navy><**b**>Tavily search **engine**

Next you will use the Tavily search engine to augment the Llama 3's responses. To create a free trial Tavily Search API, sign in with your Google or Github account here:https://app.tavily.com/sign-in

```python
from tavily import TavilyClient

TAVILY_API_KEY = getpass()
tavily = TavilyClient(api_key=TAVILY_API_KEY)
```

<font color=Navy><b>Let's do a live web search on "Llama 3 fine-tuning" using Tavily

```python
response = tavily.search(query="todays Bollywood updates")
context = [{"url": obj["url"], "content": obj["content"]} for obj in response['results']]
```

```python
context
```

<font color=Navy><b>Create documents based on the search results, index and save them to a vector store, then create a query engine.

```python

from llama_index.core import Document

documents = [Document(text=ct['content']) for ct in context]
index = VectorStoreIndex.from_documents(documents)

query_engine = index.as_query_engine(streaming=True)
```

ask Llama 3 questions about the live data using the query engine.

```python
response = query_engine.query("give me a summary")
response.print_response_stream()
```
